```python
os.makedirs(output_folder, exist_ok=True)

for filename in sorted(os.listdir(dicom_dir)):
    filepath = os.path.join(dicom_dir, filename)
    try:
        ds = dcmread(filepath)
        img = Image.fromarray(ds.pixel_array)
        img = img.convert("L")  # grayscale
        img.save(os.path.join(output_folder, f"{filename}.jpg"))
    except Exception as e:
        print(f"Skipped {filename}: {e}")
```

```python
from PIL import Image
import os
from natsort import natsorted

jpg_folder = r'D:\projects\convert_images'
output_pdf = r'D:\projects\output.pdf'

# Natural sort (IM1, IM2, ..., IM10, IM11...)
jpg_files = natsorted([
    os.path.join(jpg_folder, f)
    for f in os.listdir(jpg_folder)
    if f.lower().endswith('.jpg')
])

images = [Image.open(f).convert('RGB') for f in jpg_files]

if images:
    images[0].save(output_pdf, save_all=True, append_images=images[1:])
```
